```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from nltk import *
from nltk.corpus import stopwords
nltk.download()
```

```python

dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/Cleaned_data.csv')
```

```python
dataset.head()
```

# Tokenization

```python
def tokenize(x):
  return nltk.word_tokenize(x)
```

```python
dataset.Review = dataset.Review.apply(tokenize)
```

```python
dataset
```

# Removing Stop Words

```python
stop_words = nltk.corpus.stopwords.words('english')
```

```python
def remove_stop_word(x):
  return [w for w in x if w not in stop_words]
```

```python
dataset.Review = dataset.Review.apply(remove_stop_word)
```

```python
dataset
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
stemmer =PorterStemmer()
```

```python
def stemming(x):
  s = [stemmer.stem(y) for y in x]
  return ' '.join(s)
```

```python
dataset.Review = dataset.Review.apply(stemming)
```
