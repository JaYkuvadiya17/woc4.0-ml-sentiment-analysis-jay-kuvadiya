#Extract numbers

```python
dataset['Unique ID'] = dataset['Unique ID'].str.extract('(\d+)').astype(int)
```

```python
dataset.head()
```

# Set Index

```python
dataset.set_index('Unique ID',inplace=True)
```

```python
dataset.dtypes
```

#Lable Encoding


```python
category_column =['Suburb','Address','Type','Method','SellerG','Date','CouncilArea','Regionname']
```

```python
for cat in category_column:
  dataset[cat] = LabelEncoder().fit_transform(dataset[cat])
```

```python
dataset.dtypes
```
